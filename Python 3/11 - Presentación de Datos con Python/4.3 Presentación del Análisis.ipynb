```python
columnas = ['accion', 'fecha', 'precio']

df = pd.read_csv('https://raw.githubusercontent.com/pat-lp/Cursos-de-Python/main/Python%203/11%20-%20Presentaci%C3%B3n%20de%20Datos%20con%20Python/dow_jones_index.data',  
                usecols = [1, 2, 6],
                header= 0,
                names = columnas,
                parse_dates= ['fecha'],
                converters = {'precio':
                              lambda x:
                              float(x.replace('$', ''))
                              }

                )

df
```

####¿Qué período de tiempo representan los datos?

```python
df.fecha.dt.year.value_counts()
```

Los datos pertenecen al año 2011.

####¿A qué meses del 2011 pertenecen los datos?

```python
df.fecha.dt.month_name().value_counts()
```

Los datos se distribuyen en el primer semestre del año.
En el mes de Abril hay 30 registros más que en el resto de los meses...¿por qué?

```python
df['mes'] = df.fecha.dt.month_name()

filtro1 = df['mes'] == 'April'
filtro2 = df['mes'] == 'May'
```

```python
df[filtro1].groupby('accion')['mes'].value_counts()
```

```python
df[filtro2].groupby('accion')['mes'].value_counts()
```

####¿Cuáles acciones en promedio tuvieron la mayor y menor cotización?

```python
df.groupby('accion').agg({'precio': [np.mean, np.std]}).sort_values([('precio', 'mean')], ascending=False)
```

En promedio IBM tuvo el mayor precio: 163,10 USD.

En promedio BAC tuvo el menor precio: 13,05 USD.

####¿Cuáles acciones tuvieron la mayor y menor variación de precio?

```python
df.groupby('accion').agg({'precio': [np.mean, np.std]}).sort_values([('precio','std')], ascending=False)
```

CAT es la acción más volátil, con una desviació típica de 6,22 USD.

AA es la acción más consistente, con una desviación típica de 0,77 USD.

Conociendo la acción más volátil, podemos preguntarnos si el precio tiene correlación con el precio de otras acciones.
Se utilizrá la función **df.pivot_table()** para generar un nuevo DataFrame reorganizado con la forma deseada.

```python
df_nuevo = df.pivot_table(index='fecha',
                          columns = 'accion',
                          values='precio')
df_nuevo
```

```python
fig, (pan1,pan2) = plt.subplots(2,1, sharex=True, constrained_layout=True)

pan1.plot(df_nuevo.index, df_nuevo.CAT, 
         color = 'blue', 
         marker = 'o', 
         linestyle = 'solid',
         label = 'CAT')


pan2.plot(df_nuevo.index, df_nuevo.AA, 
          color = 'orange', 
          marker = 'v', 
          linestyle = 'solid',
          label='AA')


for pan in fig.get_axes():
    pan.set(xlabel='Fecha', 
            ylabel='Dólares americanos.', 
            title='Precio de la acción (al cierre).')

pan1.legend()
pan2.legend()
```

Los precios de ambas acciones presentan un comportamiento alcista hasta Abril -con un ligero declive en las primeras semanas de Marzo-, a partir de ahí el precio baja hasta la última semana de Junio.
