#**4.2 Tipos de Gráficos**

```python
import numpy as np
import matplotlib.pyplot as plt
```

####Gráfico de líneas

```python
año = [1950, 1960, 1970, 1980, 1990, 2000, 2010]
PIB = [300.2, 543.3, 1075.9, 2862.5, 5979.6, 10289.7, 14958.3]

plt.plot(año, PIB, color='green', marker ='o', linestyle = 'solid')

plt.xlabel('Año')
plt.ylabel('PIB')
plt.title('Gráfico PIB Nominal')
```

####Scatter Plot

```python
x1 = [3, 5, 7]
y1 = [3, 5, 9]

x2 = [1, 2, 3, 4]
y2 = [2, 3, 2, 3]

x3 = [8, 7, 2, 5, 4]
y3 = [6, 8, 7, 8, 7]

plt.scatter(x1, y1, marker='o', color ='blue')
plt.scatter(x2, y2, marker= '+', color = 'green')
plt.scatter(x3, y3, marker = '*', color = 'orange')

plt.title("Ejemplo gráfico de Scatter Plot")
plt.show()
```

####Gráfico de Barras

```python

x1 = [1, 3, 4, 5, 6, 7, 9]
y1 = [4, 7, 7, 4, 7, 8, 3]

x2 = [2, 4, 6, 8, 10]
y2 = [5, 6, 2, 6, 2]

plt.bar(x1, y1, color= 'orange', label='x1,y1')
plt.bar(x2, y2, color= 'blue', label = 'x2, y2')


plt.xlabel('x')
plt.ylabel('y')
plt.legend(loc='best')
plt.title('Gráfico de Barras')
plt.show()
```

####Histogramas

```python
y = 5 + np.random.randn(100)

x = [e for e in range(len(y))]

plt.bar(x, y, color='green')
plt.xlabel('x')
plt.ylabel('y')
plt.title("Histograma")

plt.show()
```

```python

plt.hist(y, bins=20, color='green')
plt.xlabel('x')
plt.ylabel('y')
plt.title("Histograma")

plt.show()
```

```python
plt.hist(y, cumulative=True, bins=20, color='green')
plt.xlabel('x')
plt.ylabel('y')
plt.title("Histograma")

plt.show()
```

####Gráficos Apilados

```python
x = [1, 2, 3, 4, 5] 
y1 = [1, 1, 2, 3, 5] 
y2 = [0, 4, 2, 6, 8] 
y3 = [1, 3, 5, 7, 9] 
  
etiquetas = ["Sector Privado", "Sector Público", "Sector Informal"] 
  
fig, panel = plt.subplots() 

panel.stackplot(x, y1, y2, y3,  
                labels = etiquetas,
                colors=['r','w','b']) 
  
panel.legend(loc ='upper left') 
  
panel.set_title('Gráfico apilado') 

plt.show() 
```

####Gráficos de Torta

```python
etiquetas = 'Financiero', 'Salud', 'Construcción'
sectores = [56, 66, 24]
colores = ['c', 'b', 'r']

plt.pie(sectores, labels=etiquetas, colors=colores,
        startangle=90,
        explode = (0, 0, 0.5),
        autopct = '%1.2f%%',
        shadow = True
        )

plt.axis('equal')
plt.title('Ejemplo Gráfico de Torta')
plt.show()
```

####Boxplot

```python
y1 = np.random.normal(100, 10, 200)
y2 = np.random.normal(80, 30, 200)
y3 = np.random.normal(90, 20, 200)
y4 = np.random.normal(70, 25, 200)

data_to_plot = [y1, y2, y3, y4]

fig, panel= plt.subplots()

panel.set_xticklabels(['Grupo_1', 'Grupo_2', 'Grupo_3', 'Grupo_4'])

panel.boxplot(data_to_plot,patch_artist=True)

plt.title('Ejemplo de Boxplot')
```
