#**4.1 Figuras y Paneles**

* Colores y estilos.
* Leyendas, etiquetas de ejes y t√≠tulos.

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
fig, panel = plt.subplots()

x = np.linspace(0, 4 * np.pi, 100)
y = np.sin(x)

panel.plot(x, y, label= 'Seno', linestyle = '--', color = 'y', marker='o', markeredgecolor = 'y', markerfacecolor = 'r' )

panel.legend(loc='best')

plt.title("Seno")
plt.tight_layout()
```

```python
#DOS FIGURAS HORIZONTALES

fig, (pan1, pan2) = plt.subplots(2, 1, sharex = True) #DOS FILAS Y UNA COLUMNA

x1 = np.linspace(0, 4 * np.pi, 100)
y1 = np.sin(x1)
pan1.plot(x1,y1)

#DOS FIGURAS HORIZONTALES

x2 = np.linspace(0, 4 * np.pi, 100)
y2 = np.cos(x2)
pan2.plot(x2,y2, 'tab:orange')

#EJES
pan2.set_xlabel('$x$')
pan1.set_ylabel('$y$')
```

```python
#DOS FIGURAS EN VERTICAL
fig, (pan1, pan2) = plt.subplots(1, 2,  sharey=True, constrained_layout = True)

x1 = np.linspace(0, 4 * np.pi, 100)
y1 = np.sin(x1)
pan1.plot(x1,y1)

#DOS FIGURAS HORIZONTALES

x2 = np.linspace(0, 4 * np.pi, 100)
y2 = np.cos(x2)
pan2.plot(x2,y2, 'tab:green')

#EJES
pan1.set_xlabel('$x$')
pan1.set_ylabel('$y$')
```

####.twinx()

Ambas funciones en un mismo panel.

```python
fig, pan_1 = plt.subplots(1, 1) #DOS FILAS Y UNA COLUMNA
pan_2 = pan_1.twinx()

x1 = np.linspace(0, 4 * np.pi, 100)
y1 = np.sin(x1)
pan_1.plot(x1,y1, 'g', label = 'Seno')
pan_1.legend(loc= 'best')

#DOS FIGURAS HORIZONTALES

x2 = np.linspace(0, 4 * np.pi, 100)
y2 = np.cos(x2)
pan_2.plot(x2,y2, linestyle = '--', color = 'orange', markeredgecolor = 'y', markerfacecolor = 'r', label='Coseno')
pan_2.legend(loc = 'best')

#EJES
pan_1.set_xlabel('$x$')
pan_2.set_ylabel('$y$')

#TITULO
plt.title("Seno y Coseno")
plt.tight_layout()
```
