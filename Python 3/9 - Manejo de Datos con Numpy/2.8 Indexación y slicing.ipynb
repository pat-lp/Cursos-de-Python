#2.8 Indexación y slicing

```python
import numpy as np
```

```python
#CREO UN ARRAY

arr = np.array(range(64)).reshape((8,8))
print("ARRAY ORIGINAL\n",arr)
```

```python
#SELECCIONO DE LA FILA 1 EL ELEMENTO 2
print("ARRAY ORIGINAL\n",arr)
print("\narr[1, 2] = ",arr[1, 2]) #10
```

```python
#SELECCIONO DESDE LAS FILAS 0 A 6, LAS FILAS DE 2 EN 2
print("arr[0: 7: 2,] ES: \n", arr[0: 7: 2,])
```

```python
#DE LA FILA 1, TOMA LOS ELEMENTOS DEL 3 HASTA EL FINAL
print("arr[1, 3:] ES \n",arr[1, 3:])
```

```python
#DE LAS FILAS 1, 3 Y 5, LOS ELEMENTOS DE LAS POSICIONES 3, 6 Y 7
print("arr[ [1, 3, 5], [3, 6, 7]] ES \n",arr[ [1, 3, 5], [3, 6, 7]])
```

```python
#DE LAS FILAS 1, 3 Y 5, TRAE TODOS LOS ELEMENTOS DE 3 EN 3
print("arr[ [1, 3, 5], ::3] ES \n",arr[ [1, 3, 5], ::3])
```

#Indexación booleana

```python
#CREO UN ARRAY
data = np.arange(8)
data
```

```python
#QUIERO SABER CON TRUE O FALSE QUE VALORES SON MENORES A 4
data < 4
```

```python
#QUE VALORES SON LOS MENORES A 4
data[data < 4]
```

```python
#QUE VALORES SON LOS MENORES A 4
data[np.array([ True,  True,  True,  True, False, False, False, False])]
```
