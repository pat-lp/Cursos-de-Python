#2.6 Agregar y quitar valores a un array

```python
import numpy as np
```

```python
arr = np.arange(0, 20, 2)
print(" ARRAY CON INICIO EN 0 HASTA 18\n",arr )
arr_flot = np.random.rand(4, 3)
print("\n ARRAY CON 4 FILAS Y 3 COLUMNAS\n",arr_flot)
arr_ent = np.random.randint(10, size=(2,3))
print("\nCREA ARRAY CON VALORES ENTEROS ENTRE 0 Y 9 INCLUSIVE\n",arr_ent)
arr_6 = np.full((3,3), 6)
print("\nMATRIZ DE 3X3, CON 6 COMO VALOR\n",arr_6)
print("\nAPPEND AGREGA VALORES AL FINAL DEL ARRAY\n",np.append(arr, [12, 13, 14, 51]))
print("\nINSERT: AGREGA VALORES EN UNA POSICIÓN ESPECÍFICA\n",np.insert(arr, 2, [42, 34]))
print("\nELIMINA LOS VALORES DEL INDICE EJE 0\n",np.delete(arr_ent, 1, axis=0))
```
