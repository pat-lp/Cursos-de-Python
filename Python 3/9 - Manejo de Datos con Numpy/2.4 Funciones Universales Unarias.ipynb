#**2.4 Funciones Universales Unarias**
Son funciones que realizan operaciones elemento a elemento sobre los datos de un ndarray, reciben un solo array como argumento.


* abs, fabs: Calculan el valor absoluto.
* sqrt: Calculan la raíz cuadrada.
* square: Calcula el cuadrado.
* exp: Calcula la exponencial.
* log, log10, log2, log1p: Calcula el logaritmo natural en base 10, 2, log(x+1).
* isnan: Devuelve un array booleano indicando si cada elemento es NaN(Not a Number).
* sin, cos, tan: Devuelve un array con el resultado de aplicar la función seno, coseno, tangente en los valores.
* mean, median, std, var: Calcula la media artimética, le mediana, la desviación estandard y la varianza de un array.



```python
import numpy as np

a = np.array([2,4,9], float)
```

```python
print(f"VALOR ABSOLUTO.........: {np.abs(a)}")
print(f"RAIZ CUADRADA..........: {np.sqrt(a)}")
print(f"EXPONENCIAL............: {np.exp(a)}")
print(f"LOGARITMO EN BASE 10...: {np.log10(a)}")
print(f"ARRAY NaN..............: {np.isnan(a)}")
print(f"SENO, COSENO Y TANGENTE: {np.sin(a)},{np.cos(a)},{np.tan(a)}")
print(f"\nFUNCIONES ESTADÍSTICAS")
print(f"MEDIA..................: {np.mean(a)}")
print(f"MEDIANA................: {np.median(a)}")
print(f"DESVIACIÓN ESTANDAR....: {np.std(a)}")
print(f"VARIANZA...............: {np.var(a)}")
```
