#2.7 Transformaciones

```python
import numpy as np
```

## As Type
Este método nos permite modificar el tipo de dato de los valores de un array.

```python
arr_ent = np.random.randint(100, size = (3, 4))
print("ARRAY CON NÚMEROS ENTEROS\n",arr_ent)
print("\nMISMO ARRAY CON NÚMEROS FLOTANTES\n",arr_ent.astype(float))
```

#Sort

Este método nos permite ordenar los valores de un array.

```python
arr_ent = np.random.randint(100, size = (3, 4))
print("ARRAY ORIGINAL\n",arr_ent)
arr_ent.sort()
print("\nARRAY ORDENADO FILA X FILA\n",arr_ent)
arr_ent.sort(axis = 0)
print("\nARRAY ORDENADO COLUMNA X COLUMNA\n",arr_ent)
```

#Reshape

Este método nos permite modificar las dimensiones de un array.

```python
arr_ent = arr_ent.reshape(4, 3) 
print(arr_ent)
```

#Flatten

Este método crea un array unidimensional desde el original.

```python
plano_arr = arr_ent.flatten()
print(" ARRAY UNIDIMENSIONAL\n",plano_arr)
```

#To List

Este método crea una lista a partir de un array.

```python
lista_arr = plano_arr.tolist()
print(" LISTA A PARTIR DE UN ARRAY\n",lista_arr)
```

#Separar y juntar arrays


```python
arrays = np.split(arr_ent, 4)
print("ARRAY ORIGINAL\n",arr_ent)
print("\nSEPARA EL ARRAY ANTERIOR EN ARRAYS SEPARADOS\n",arrays)

print("\nCONCATENA LOS ARRAYS SEPARADOS DE POSICIÓN 1 CON EL DE POSICIÓN 0\n",np.concatenate((arrays[1], arrays[0]), axis = 0))
```
