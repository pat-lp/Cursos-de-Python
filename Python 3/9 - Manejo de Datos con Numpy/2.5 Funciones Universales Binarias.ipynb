#2.5 Funciones Universales Binarias

Son funciones que realizan operaciones elemento a elemento del ndarray, toman dos arrays y devuelven uno ó más arrays.

* add: Suma entre elementos.
* subtract: Resta entre elementos.
* multiply: Mutiplica los arrays.
* divide, floor_divide: Divide ó división truncada.
* array_equal: Devuelve True si los elementos del array son iguales entre sí.
* power: Eleva cada elemento a la potencia indicada.
* fmin: Devuelve el mínimo.
* fmax: Devuelve el máximo.

```python
import numpy as np

array1 = np.array([5 ,36, 17, 18, 9])
array2 = np.array([8, 24, 17, 19, 9])

print(f"SUMA............: {np.add(array1,array2)}")
print(f"RESTA...........: {np.subtract(array1,array2)}")
print(f"MULTIPLICACIÓN..: {np.multiply(array1,array2)}")
print(f"DIVISIÓN........: {np.divide(array1,array2)}")
print(f"SON IGUALES?....: {np.array_equal(array1,array2)}")
print(f"POTENCIA........: {np.power(array1,array2)}")
print(f"MÍNIMO..........: {np.fmin(array1,array2)}")
print(f"MÁXIMO..........: {np.fmax(array1,array2)}")
```
