#**Series**

Las Series contienen un array unidimensional de datos y un array de índices.

```python
import numpy as np
import pandas as pd
```

```python
#Se crea una Serie

serie = pd.Series([1990, 1991, 1992])

print(f"SERIE\n{serie}\n")

#VALUES: Para imprimir los valores de la Serie
print(f"VALORES DE LA SERIE\n{serie.values}\n")

#VALUES: se pueden realizar otras operaciones
print(f"SUMAR VALORES DE LA SERIE\n{serie.values.sum()}\n")

#INDEX: Indices de la Serie
print(f"ÍNDICES DE LA SERIE\n{serie.index}\n")

```

```python
#SE DETERMINA UN INDICE PARA LA SERIE, PRIMERO SON LOS VALORES DE LA SERIE Y LUEGO LOS ÍNDICES

serie_indice = pd.Series ([1990, 1991, 1992], index=['Ana', 'Megan', 'Mariana'])
print(f"SERIE CON ÍNDICE\n{serie_indice}")
```

```python
#CREAR UNA SERIE A PARTIR DE UN DICCIONARIO

serie_diccionario = pd.Series({"Cuadrado de {}".format(i): i**2 for i in range(1,11)})

print(serie_diccionario)
```
