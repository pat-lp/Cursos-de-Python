#**3.6 Fechas con Pandas**

```python
import datetime
import numpy as np
import pandas as pd
```

####**.to_datetime()**

Convierte una columna de un DataFrame en un objeto timestams de Pandas.

```python
fecha = pd.to_datetime('23FEB2021')

fecha.year
```

####**.date_range()**

Permite generar un rango de fechas.

```python
fecha_inicio = pd.date_range(start = '24/4/2022', end = '24/5/2022', freq='D')

print(f"FECHA INICIO \n{fecha_inicio}")

fecha_fin = pd.date_range(start = '24/5/2022', end = '24/6/2022', freq='D')

print(f"FECHA FIN \n{fecha_fin}")
```

```python
lista_random = []

for i in range(15):
  lista_random.append(np.random.randint(2))

#DATAFRAME
df = pd.DataFrame()
df['INICIO_CAMPAÑA'] = fecha_inicio[:15]
df['FIN_CAMPAÑA'] = fecha_fin[:15]
df['TARGET'] = lista_random


df
```

```python
df['DIA DE INICIO'] = df['INICIO_CAMPAÑA'].dt.day

df
```
