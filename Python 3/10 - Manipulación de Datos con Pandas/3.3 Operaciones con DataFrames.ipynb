#**3.3 Operaciones con DataFrames**

```python
import numpy as np
import pandas as pd
```

```python
nros_array = np.random.randint(1, 20, size=10, dtype=int) #CREO UN ARRAY CON VALORES RANDOM

df = pd.DataFrame(nros_array)
df
```

```python
#AGREGAR UNA NUEVA COLUMNA

df['valor_inicial'] = 10

df['perdidas'] = [(i+2)*np.e for i in range(10)]

df['total'] = round(df['perdidas'] * 100, 2)

df['Columna_random'] = 0
df
```

```python
#CAMBIAR EL NOMBRE DE UNA COLUMNA, PRIMERO IMPRIMO LOS NOMBRES DE LAS COLUMNAS

df.columns
```

```python
#MODIFICO LOS NOMBRES DE LAS COLUMNAS

df.columns = ['id', 'Valor_inicial', 'Perdidas', 'Total', 'Columna_random']
df
```

```python
#MODIFICO UNA COLUMNA

df['Valor_inicial'] = [(i*8.5)+i for i in range(1,11)]

df
```

```python
#ELIMINAR UNA COLUMNA

del df['Columna_random']

df
```

```python
#ELIMINAR FILAS Y COLUMNAS CON DROP, SIN MODIFICAR EL DF ORIGINAL

c = df.drop([2,3,4], axis = 0)
print(f"Se eliminan las filas 2, 3, 4\n{c}\n")

d = df.drop(['Valor_inicial'], axis = 1)
print(f"Se elimina la columna Valor_inicial\n{d}\n")
```

```python
#MODIFICAR EL ÍNDICE

i = ['001', '002', '003', '004', '005', '006', '007', '008', '009', '010']

df.index = i

df
```
