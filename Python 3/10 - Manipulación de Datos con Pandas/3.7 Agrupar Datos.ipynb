#**3.7 Agrupar Datos**

```python
import numpy as np
import pandas as pd
```

```python

data = pd.read_csv('https://raw.githubusercontent.com/pat-lp/Cursos-de-Python/main/Python%203/10%20-%20Manipulaci%C3%B3n%20de%20Datos%20con%20Pandas/data_celular.csv',
                   header = 0, #COLUMNAS
                   index_col = 0, #COLUMNA TOMADA COMO INDICE
                   names = ['indice', 'fecha','duracion', 'item', 'mes', 'red', 'tipo_red'],
                   parse_dates = ['fecha']
                   )

data.head()
```

```python
#INFORMACION DEL DATAFRAME

print(f"CUANTAS FILAS TIENE EL DATAFRAME: {data['item'].count()}")
print(f"TIEMPO TOTAL EN SEGUNDOS REGISTRADO EN LLAMADAS: {data['duracion'][data['item'] == 'call'].sum()}")
print(f"CON CUANTAS REDES TELEFONICAS SE CONTACTO EN EL PERIODO 2014/11 AL 2015/03: {data['red'].nunique()}")
```

####**.groupby()**

Permite agrupar datos para obtener informaci√≥n.

```python
data.groupby('mes').groups. values()
```

```python
print("CANTIDAD DE ENTRADAS POR MES SEGREGADAS EN LLAMADAS, SMS Y DATOS")
print(data.groupby(['mes','item'])['duracion'].count())
```

```python
print("DURACION TOTAL DE LLAMADAS REALIZADAS A CADA UNA DE LAS OPERADORES ES:")
data[data['item'] == 'call'].groupby('red')['duracion'].sum()
```

```python
print("CUANTAS LLAMADAS, SMS Y DATOS SON ENVIADOS A CADA OPERADORA POR MES")
data.groupby(['mes','tipo_red'])['fecha'].count()
```
