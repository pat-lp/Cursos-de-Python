#**3.5 Fechas con DateTime**

```python
from datetime import date

fecha1 = date(2022, 1, 25)
print(fecha1)
print(f"DIA: {fecha1.day}")
print(f"MES: {fecha1.month}")
print(f"AÑO: {fecha1.year}")


print(f"La fecha de hoy: {date.today()}")
```

```python
import datetime
#DATETIME.TIME

from datetime import time

hora1 = time(10, 23, 56, 7)
print(hora1)

print(f"HORA: {hora1.hour}")
print(f"MINUTOS: {hora1.minute}")
print(f"SEGUNDOS: {hora1.second}")
print(f"MICROSEGUNDOS: {hora1.microsecond}")
print(f"Día y Hora actual: {datetime.datetime.now()}")
```

####.weekday(): Devuelve un número correspondiente al día de la semana.
Lunes: 0, Martes:1, Miércoles:2, Jueves:3, Viernes:4, Sábado:5, Domingo:6

```python
fecha1.weekday()
```

####.isoweekday(): Devuelve un número correspondiente al día de la semana, pero comienza por el Domingo.

Domingo:0, Lunes: 1, Martes:2, Miércoles:3, Jueves:4, Viernes:5, Sábado:6 

```python
fecha1.isoweekday()
```

####**.isocalendar()**

Devuelve una tupla que representa una fecha:
- valor_1: Indica el año en formato ISO.
- valor_2: Indica la semana en formato ISO.
- valor_3: Indica el día en formato ISO.

```python
fecha1.isocalendar()
```

```python
#CUANDO SE TIENE UNA FECHA EN FORMATO STRING

f1_datetime = date.fromisoformat('2022-06-28')

print(f1_datetime)
print(type(f1_datetime))

f2_ISO = date(2022, 6, 18).isoformat()

print(f2_ISO)
print(type(f2_ISO))
```

####**.strptime()**

Permite leer una cadena de texto con fecha como un objeto de clase *.datetime*.


|Formatos|Descripción|
|-------|-----------|
|`%a`|Nombre del día de la semana abreviado|
|`%A`|Nombre del día de la semana completo|
|`%w`|Día de la semana como número decimal|
|`%d`|Día del mes como número de dos dígitos|
|`%-d`|Día del mes como número decimal|
|`%b`|Nombre del mes abreviado|
|`%B`|Nombre del mes completo|
|`%m`|Mes del año número de dos dígitos|
|`%-m`|Mes del año como número decimal|
|`%y`|Año sin centenas como número de dos dígitos|
|`%-y`|Año sin centenas como número decimal|
|`%Y`|Año completo|
|`%H`|Hora en 24 hrs. con dos dígitos|
|`%-H`|Hora en 24 hrs. como número decimal|
|`%I`|Hora en 12 hrs. con dos dígitos|
|`%-I`|Hora en 12 hrs. como número decimal|
|`%M`|Minutos con dos dígitos|
|`%-M`|Minutos como número decimal|
|`%S`|Segundos con dos dígitos|
|`%-S`|Segundos como número decimal|

```python
fecha_1 = '23FEB2021:00:00:00'

dia = datetime.datetime.strptime(fecha_1,'%d%b%Y:%H:%M:%S')
dia
```

####**datime.timedelta**

Nos permite operar entre fechas

```python
d1 = datetime.datetime.now() #fecha actual
print(f"FECHA ACTUAL: {d1}")
d2 = d1 + datetime.timedelta(days = 2) #a la fecha de hoy se le suman dos dias
print(f"SE SUMAN DOS DIAS A LA FECHA DE HOY: {d2}")

d3 = d1 + datetime.timedelta(weeks = 2) #a la fecha de hoy se le suman dos dias
print(f"SE SUMAN DOS SEMANAS A LA FECHA DE HOY: {d3}")

d4 = d1 - datetime.timedelta(days = 2)
print(f"SE RESTAN DOS DIAS A LA FECHA DE HOY: {d4}")

d5 = d1 - datetime.timedelta(weeks = 2)
print(f"SE RESTAN DOS SEMANAS A LA FECHA DE HOY: {d5}")
```
