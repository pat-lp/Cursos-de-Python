#**3.4 Indexación y Filtrado**

```python
import numpy as np
import pandas as pd
```

```python
df = pd.DataFrame(
    np.random.randint(0, 10, size=(10,2)),
    index = ['a', 'b', 'c', 'd', 'e', 'f', 'g', 'h', 'i', 'j'],
    columns = ['Cod_Empleado', 'Calificación']
)

df
```

```python
#SELECCIONAR FILAS CON "LOC"

print(f"Selecciona una fila\n{df.loc['a']}\n")

print(f"Selecciona un conjunto de filas \n{df.loc['c':'f']}\n")

print(f"Selecciona filas de una columna en particular\n{df.loc[df.index[3:7],'Cod_Empleado']}\n")
```

```python
#SELECCIONAR FILAS CON "ILOC"

print(f"Se selecciona hasta la fila 3 \n{df.iloc[:4]}\n")

print(f"Se selecciona de la fila 4 en adelante \n{df.iloc[4:]}\n")
```
